```python
#請取出第一個地區所記錄的每一個時間點跟溫度
import xmltodict
import xml.dom.minidom

#利用xmltodict讀取到第一個地區
with open("E:\\Python\\GitHub\\Day003_example\\64_72hr_CH.xml", "r",encoding="utf-8") as fd:
    doc = xmltodict.parse(fd.read())

loc = doc["cwbopendata"]['dataset']['locations']['location'][0]
print(loc['locationName'])

#利用minidom讀取該地區所有時間及溫度資料
def getText(nodelist):
    rc = ""
    if nodelist.nodeType == nodelist.ELEMENT_NODE:
        rc = rc + nodelist.getElementsByTagName('value')[0].firstChild.nodeValue
    return rc

doc2 = xml.dom.minidom.parse("E:\\Python\\GitHub\\Day003_example\\64_72hr_CH.xml")
root = doc2.documentElement 

# get first children elements: <weatherElement/> <weatherElement/> 
weather = root.getElementsByTagName("weatherElement")[0]
timeNode = weather.getElementsByTagName("time")
for timeValue in timeNode:
    datetime = timeValue.getElementsByTagName("dataTime")[0]
    weathervalue = timeValue.getElementsByTagName("elementValue")[0]
    print ('dataTime: ' + datetime.firstChild.data + ' ,value: ' + getText(weathervalue))
```
