```python
#YAHOO電影爬蟲練習
import requests
import re
from bs4 import BeautifulSoup

url = 'https://movies.yahoo.com.tw/movie_intheaters.html'
resp = requests.get(url)
print(resp.text)
```

```python
soup = BeautifulSoup(resp.text, 'lxml')
html = soup.find("div", attrs={'class':'release_box'})  # 尋找正在上映中的全部電影筆數
print(html.p.string)
```

```python
#爬取所有上映中的電影資料：電影名稱、期待度、滿意度、簡介等等。

#1. 點選網頁最底部的下一頁，觀察其網址變化。然後我們把頁數當作變數傳入，用迴圈爬取。
max_page = 9
for page_number in range(1, max_page+1, 1):
    url = 'https://movies.yahoo.com.tw/movie_intheaters.html'
    payload = {'page':str(page_number)}
    resp = requests.get(url, params=payload)
    resp.encoding = 'utf-8'
    soup = BeautifulSoup(resp.text, 'lxml')
    
    # 把電影介紹擷取出來
    movie_list = soup.find("ul", attrs={"class":"release_list"}).find_all("li")
    
    for p in movie_list:
        movie_name = p.find("div", attrs={"class":"release_movie_name"})
        print("電影名稱: " + movie_name.a.string, end=" ")

        level_box = movie_name.find("dl", attrs={"class":"levelbox"})
        
        level = level_box.find("dt")
        expectation = level.find("div", attrs={"class":"leveltext"})
        print("期待度：", expectation.span.string, end=" ")

        level2 = level_box.find("dd")
        satisfaction = level2.find("div", attrs={"class":"leveltext"})
        print("滿意度：", satisfaction.span["data-num"], end=" ") 
        
        release_text = p.find("div", attrs={"class":"release_text"})
        print("簡介: " + release_text.span.string, end=" ")

        print("==================================================", end="\n")
```
