```python
#利用開發者工具去尋找「 https://www.zhihu.com/explore 」和「 http://aicoin.cn 」網站的 HTTP 網址，
#以及他所帶的 Headers 分別是哪些，試著用 request 去發送有沒有帶 Headers 的差異。
import requests
url = 'https://www.zhihu.com/explore'
headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get(url,headers=headers)
r.encoding = 'utf-8'
print(r.text)
```

```python
import requests

url = 'http://aicoin.cn/'
headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get(url,headers=headers)
r.encoding = 'utf-8'

print(r.text)
```
