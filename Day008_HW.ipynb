```python
#為什麼要使用 BeatifulSoup 處理==>文件格式化
import requests
from bs4 import BeautifulSoup


r = requests.get('https://www.dcard.tw/f')
soup = BeautifulSoup(r.text, features="lxml")

mydivs = soup.findAll('div',class_='ForumEntryGroup_listContainerInner_2ZSHKa')
print(mydivs)
```

```python
#觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決==>加上header
import requests
from bs4 import BeautifulSoup

headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get('https://www.zhihu.com/explore',headers=headers)
soup = BeautifulSoup(r.text, features="lxml")

print(soup)
```
