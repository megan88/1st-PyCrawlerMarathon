```python
#哪一個類別的文章最多
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

articles = []
columns = ['date', 'tag','text']

for d in soup.find(class_="part_list_2").find_all('h3'):
    #print(d.find('em').text)
    articles.append([d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text])
    
df = pd.DataFrame(articles, columns=columns)

df = pd.DataFrame(df.groupby('tag').size(), columns=['article count'])
df['rank'] = df.rank(method='min', ascending=False)
df.sort_values(by=['rank'])
```
