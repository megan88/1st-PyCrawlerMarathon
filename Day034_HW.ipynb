```python
#寫一個可自動化抓取可用 Proxy 的 proxy_ips
from bs4 import BeautifulSoup
import requests
import random

proxy_ips = []

url = 'https://free-proxy-list.net/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

for tr in soup.find('tbody').find_all('tr'):
    proxy = tr.find_all('td')[0].text + ':' + tr.find_all('td')[1].text   #擷取ip address和port
    proxy_ips.append(proxy)
    
print(proxy_ips)
```

```python
#將免費的代理伺服器，發送至 http://ip.filefab.com/index.php 找出可用的
available_proxies = []

for i in range(len(proxy_ips)):
    ip = proxy_ips[i]
    print('Use', ip)
    try:
        resp = requests.get('http://ip.filefab.com/index.php',
                        proxies={'http': ip, 'https': ip}, timeout=10)
        soup = BeautifulSoup(resp.text, 'html5lib')
        print(soup.find('h1', id='ipd').text.strip())
        available_proxies.append(ip)
        
    except:
        print('Fail')
    
print(available_proxies)
```
